<h2 id="Black-Scholes-Model">Black Scholes Model<a class="anchor-link" href="#Black-Scholes-Model"></a></h2>



<p>The <em>Black Scholes</em> model was published in 1973 for non-dividend paying stocks and since then the model has created a revolution in quantitative finance and opened up derivatives pricing paradigm. Black Scholes model is based on number of assumptions about how financial markets operate and those are:</p>
<ul>
<li>Arbitrage Free Markets</li>
<li>Frictionless and Continuous Markets</li>
<li>Risk Free Rates</li>
<li>Log-normally Distributed Price Movements</li>
<li>Constant Volatility</li>
</ul>
<p>These assumptions maynot hold true in reality, but are not particularly limiting. The generalized Black Scholes framework have been extended to price derivaties of other asset classes such as Black 76 (Commodity Futures) and Garman-Kohlhagen (FX Futures) that are currently used in derivative pricing and risk management.</p>



<h2 id="Black-Scholes-Formula">Black Scholes Formula<a class="anchor-link" href="#Black-Scholes-Formula"></a></h2><p>The Black Scholes equation describes the price of the option over time as<br/><br/></p>
\begin{equation}
\frac{\partial V}{\partial t}
+ \frac 1{2}{\sigma^2 S^2} \frac{\partial^2 V}{\partial S^2}
+ r S \frac{\partial V}{\partial S} - rV = 0 
\label{eq:1}
\end{equation}<p><br/>Solving the above equation, we know that the value of a call option for a non-dividend paying stock is:<br/><br/></p>
\begin{equation}
    \mathrm C(\mathrm S,\mathrm t)= \mathrm N(\mathrm d_1)\mathrm S - \mathrm N(\mathrm d_2) \mathrm K \mathrm e^{-rt}
    \label{eq:2}
\end{equation}<p>and, the corresponding put option price is:<br/></p>
\begin{equation}
    \mathrm P(\mathrm S,\mathrm t)= \mathrm N(-\mathrm d_2) \mathrm K \mathrm e^{-rt} - \mathrm N(-\mathrm d_1)\mathrm S
    \label{eq:3}
\end{equation}<p>where, \begin{equation}
    \mathrm d_1= \frac{1}{\sigma \sqrt{\mathrm t}} \left[\ln{\left(\frac{S}{K}\right)} + \left(r + \frac{\sigma^2}{2} \right)t \right]
\end{equation}</p>
<p>and, \begin{equation}
    \mathrm d_2= \mathrm d_1 - {\sigma \sqrt{\mathrm t}}
\end{equation}
<br/>
\begin{equation}
    N(x)=\frac{1}{\sqrt{2\pi}} \int_{-\infty}^{x} \mathrm e^{-\frac{1}{2}z^2} dz
    \label{eq:5}
\end{equation}</p>
<p>where,</p>
<p>$S$ is the spot price of the underlying asset<br/>
$K$ is the strike price<br/>
$r$ is the annualized continuous compounded risk free rate<br/>
$\sigma$ is the volatility of returns of the underlying asset<br/>
$t$ is time to maturity (expressed in years)<br/>
$N(x)$ is the standard normal cumulative distribution function<br/></p>

<h2 id="Example">Example<a class="anchor-link" href="#Example"></a></h2><p>We will now price a European call option of an asset maturing in the next 144 days with the following specifications:</p>
<ul>
<li>$S$ = 330</li>
<li>$K$ = 335</li>
<li>$r$ = 2%</li>
<li>$\sigma$ = 20%</li>
<li>$t$ = 0.3945</li>
</ul>
<p><em>What price should you pay (per unit) for the option contract?</em></p>
<p>$\mathrm d_1= \frac{1}{0.20 \sqrt{0.3945}} \left[\ln{\left(\frac{330}{335}\right)} + 0.3945\left(0.02 + \frac{0.20^2}{2} \right) \right]=0.00590775$</p>
<p>$\mathrm d_2= d_1 - {0.20 \sqrt{0.3945}}=-0.11971071$</p>
<p>Now that we've calculated $d_1$ and $d_2$, we then evaluate
$N(d_1)=N(0.00590775)=0.50235683$</p>
<p>$N(d_2)=N(-0.11971071)=0.45235615$</p>
<p>Next, we plug these values in the above equation to calculate the price of the call option.</p>
<p>$C(330,0.3945)= (0.50234683)(330) - (0.45235615)(335) \space \mathrm e^{-(0.02)(0.3945)} \approx 15.4294$</p>
<p>Thus, we will be paying around \$1542.94 for the options contract with a lot size or multiplier of 100 units.</p>



<h2 id="Implementation">Implementation<a class="anchor-link" href="#Implementation"></a></h2>


```python

# Import required libraries
import pandas as pd
from numpy import *
from scipy.stats import norm

import matplotlib.pyplot as plt
from tabulate import tabulate

```


<h3 id="Option-Object">Option Object<a class="anchor-link" href="#Option-Object"></a></h3><p>We will construct an options class object</p>


```python

class BS:
    
    """
    This is a class for Options contract for pricing European options on stocks without dividends.
    
    Attributes: 
        spot          : int or float
        strike        : int or float 
        rate          : float
        dte           : int or float [days to expiration in number of years]
        volatility    : float
    """    
    
    def __init__(self, spot, strike, rate, dte, volatility):
        
        # Spot Price
        self.spot = spot
        
        # Option Strike
        self.strike = strike
        
        # Interest Rate
        self.rate = rate
        
        # Days To Expiration
        self.dte = dte
        
        # Volaitlity
        self.volatility = volatility
       
        # Utility 
        self._a_ = self.volatility * self.dte**0.5
        
        if self.strike == 0:
            raise ZeroDivisionError('The strike price cannot be zero')
        else:
            self._d1_ = (log(self.spot / self.strike) + \
                     (self.rate + (self.volatility**2) / 2) * self.dte) / self._a_
        
        self._d2_ = self._d1_ - self._a_
        
        self._b_ = e**-(self.rate * self.dte)
        
        
        # The __dict__ attribute
        '''
        Contains all the attributes defined for the object itself. It maps the attribute name to its value.
        '''
        for i in ['callPrice', 'putPrice', 'callDelta', 'putDelta', 'callTheta', 'putTheta', \
                  'callRho', 'putRho', 'vega', 'gamma']:
            self.__dict__[i] = None
        
        [self.callPrice, self.putPrice] = self._price()
        [self.callDelta, self.putDelta] = self._delta()
        [self.callTheta, self.putTheta] = self._theta()
        [self.callRho, self.putRho] = self._rho()
        self.vega = self._vega()
        self.gamma = self._gamma()
        
    
    # Option Price
    def _price(self):
        '''Returns the option price: [Call price, Put price]'''

        if self.volatility == 0 or self.dte == 0:
            call = maximum(0.0, self.spot - self.strike)
            put = maximum(0.0, self.strike - self.spot)
        else:
            call = self.spot * norm.cdf(self._d1_) - self.strike * e**(-self.rate * \
                                                                       self.dte) * norm.cdf(self._d2_)

            put = self.strike * e**(-self.rate * self.dte) * norm.cdf(-self._d2_) - \
                                                                        self.spot * norm.cdf(-self._d1_)
        return [call, put]

    # Option Delta
    def _delta(self):
        '''Returns the option delta: [Call delta, Put delta]'''

        if self.volatility == 0 or self.dte == 0:
            call = 1.0 if self.spot > self.strike else 0.0
            put = -1.0 if self.spot < self.strike else 0.0
        else:
            call = norm.cdf(self._d1_)
            put = -norm.cdf(-self._d1_)
        return [call, put]

    # Option Gamma
    def _gamma(self):
        '''Returns the option gamma'''
        return norm.pdf(self._d1_) / (self.spot * self._a_)

    # Option Vega
    def _vega(self):
        '''Returns the option vega'''
        if self.volatility == 0 or self.dte == 0:
            return 0.0
        else:
            return self.spot * norm.pdf(self._d1_) * self.dte**0.5 / 100

    # Option Theta
    def _theta(self):
        '''Returns the option theta: [Call theta, Put theta]'''
        call = -self.spot * norm.pdf(self._d1_) * self.volatility / (2 * self.dte**0.5) - self.rate * self.strike * self._b_ * norm.cdf(self._d2_)

        put = -self.spot * norm.pdf(self._d1_) * self.volatility / (2 * self.dte**0.5) + self.rate * self.strike * self._b_ * norm.cdf(-self._d2_)
        return [call / 365, put / 365]

    # Option Rho
    def _rho(self):
        '''Returns the option rho: [Call rho, Put rho]'''
        call = self.strike * self.dte * self._b_ * norm.cdf(self._d2_) / 100
        put = -self.strike * self.dte * self._b_ * norm.cdf(-self._d2_) / 100

        return [call, put]

```

```python

# Initialize option
option = BS(100,100,0.0,1,0.2)

header = ['Option Price', 'Delta', 'Gamma', 'Theta', 'Vega', 'Rho']
table = [[option.callPrice, option.callDelta, option.gamma, option.callTheta, option.vega, option.callRho]]

print(tabulate(table,header))

```


<h3 id="Visualize">Visualize<a class="anchor-link" href="#Visualize"></a></h3><p>Plot option values for range of spot prices over differece time</p>


```python

figure, axes = plt.subplots(1,2, figsize=(20,6), sharey=True)

x = linspace(50,150,100)#
d = {'0M':1e-50, '3M': 0.25, '6M': 0.5, '9M': 0.75, '12M': 1.0}

for k,v in d.items():
    axes[0].plot(x, BS(x,100,0.0,v,0.2).callPrice, label=k)
    axes[1].plot(x, BS(x,100,0.0,v,0.2).putPrice, label=k)
    
# Set axis title
axes[0].set_title('Call Price'), axes[1].set_title('Put Price')

# Define legend
axes[0].legend(), axes[1].legend()

plt.show()

```

```python

# Plot straddle price 
plt.plot(x, (BS(x,100,0.0,1,0.2).callPrice + BS(x,120,0.0,1,0.2).putPrice))
plt.title('Straddle Price');

```

```python

# Plot graph iteratively
fig, ax = plt.subplots(2,2, figsize=(20,10))

for k,v in d.items():
    ax[0,0].plot(x, BS(x,100,0.0,v,0.2).callDelta, label=k)
    ax[0,1].plot(x, BS(x,100,0.0,v,0.2).gamma, label=k)
    ax[1,0].plot(x, BS(x,100,0.0,v,0.2).vega, label=k)
    ax[1,1].plot(x, BS(x,100,0.0,v,0.2).callTheta, label=k)
    
# Set axis title
ax[0,0].set_title('Delta'), ax[0,1].set_title('Gamma'), ax[1,0].set_title('Vega'), ax[1,1].set_title('Theta')

# Define legend
ax[0,0].legend(), ax[0,1].legend(), ax[1,0].legend(), ax[1,1].legend()

plt.show()

```


<h2 id="SPY-Option">SPY Option<a class="anchor-link" href="#SPY-Option"></a></h2><p>Let's now retrieve SPY option price from Yahoo Finance using <code>yfinance</code> library and manipulated the dataframe using the above Black Scholes option pricing model that we created.</p>
<p><a href="https://finance.yahoo.com/quote/SPY/options?date=1609372800&amp;p=SPY">https://finance.yahoo.com/quote/SPY/options?date=1609372800&amp;p=SPY</a></p>


```python

import yfinance as yf

```

```python

# Get SPY option chain
spy = yf.Ticker('SPY')
options = spy.option_chain('2022-04-08')

```

```python

# Dec 2020 335 SPY call option price
spot = 335
strike = 335
rate = 0.0
dte = 0.3945
vol = 0.2185
spy_opt = BS(spot,335,rate,dte,vol)
print(f'Option Price of SPY201231C00335000 with BS Model is {spy_opt.callPrice:0.4f}')

```

```python

# Filter calls for strike above 330
df = options.calls[options.calls['strike']>330]
df.reset_index(drop=True, inplace=True)

# Check the filtered output
df.iloc[:,:11].head()

```

```python

# Dataframe manipulation with selected fields
df = pd.DataFrame({'Strike': df['strike'], 
                   'Price': df['lastPrice'], 
                   'ImpVol': df['impliedVolatility']})
# Check output
df.head(2)

```

```python

# Derive greeks and assign to dataframe as columns
df['Delta'] = df['Gamma'] = df['Vega'] = df['Theta'] = 0.

for i in range(len(df)):
    
    df['Delta'].iloc[i] = BS(spot,df['Strike'].iloc[i],rate,dte,df['ImpVol'].iloc[i]).callDelta
    df['Gamma'].iloc[i] = BS(spot,df['Strike'].iloc[i],rate,dte,df['ImpVol'].iloc[i]).gamma
    df['Vega'].iloc[i] = BS(spot,df['Strike'].iloc[i],rate,dte,df['ImpVol'].iloc[i]).vega
    df['Theta'].iloc[i] = BS(spot,df['Strike'].iloc[i],rate,dte,df['ImpVol'].iloc[i]).callTheta

```
